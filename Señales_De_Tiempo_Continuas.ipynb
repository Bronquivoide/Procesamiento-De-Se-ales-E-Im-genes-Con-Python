### En este notebook se pretende simular la generación señales (mostrarlas gráficamente) que son ejemplos básicos de señales continuas en el tiempo; es decir, señales del tipo $x(t)$ tal que $t$ toma infinitos valores en un intervalo $[t_1,t_2]$. 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import scipy
from scipy import signal
```

# Ejemplos de generación de señales de tiempo continuo:

# $\blacksquare$ Señal impulso unitario:

### Este tipo de señal se trata analíticamente mediante una función que la describe, la cual está definida como una delta de Dirac:

## $ \delta(t)= \begin{cases}
    1,& t=0\\
    0,              & t \neq 0
\end{cases}$

### Nótese que aunque la función es discontinua (hablando en términos del análisis real), en realidad la señal es continua en el tiempo.

### Aunque matemáticamente la función se vea gráficamente de la siguiente forma...

```python
t=np.arange(-5, 6, 1)
delta=np.zeros_like(t)
delta[t == 0]=1
markerline, stemlines, baseline =plt.stem(t,delta,"b")
plt.title(r'$\delta(t)$')
plt.xlabel('t')
plt.ylabel('$\delta(t)$')
plt.setp(baseline,color='b', linewidth=1)
plt.grid(False)
plt.show()
```

### en realidad, en un sentido físico la magnitud de la señal iría aumentando en una vecindad cercana (de valores $t<0$) a $t=0$. Tomamos que cuando $t=-1$, la magnitud de la señal empieza a aumentar hasta alcanzar su valor máximo en $t=0$. Es decir, se empezaría a notar una pendiente creciente cuando $-1 \leq t <0$.

### La señal se vería así (dándole un formato al mejor estilo osciloscopio):

```python
impulso=signal.unit_impulse(20,"mid")
print(impulso)
plt.figure(figsize = (18,8))
plt.title("Impulso unitario", fontsize = 40, color="yellow")
style.use('dark_background') 
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(np.arange(-10, 10), impulso, linewidth = 3,color="cyan")
plt.xlabel('$t$', fontsize = 30,color ="yellow")
plt.ylabel('$\delta (t)$, amplitud', fontsize = 30,color ="yellow")
plt.grid(True)
plt.show()
```

### *Nota: La primera gráfica que mostramos de $\delta (t)$ puede verse como la descripción discreta de la señal tipo impulso unitario.*

### Podemos también tomar un retraso y un avance temporal en la señal, lo que se conoce como señal a tiempo cambiado (tal vez la mejor traducción para *"shifted time signal"*).

### Cuando tenemos una señal de tiempo continuo $x(t)$, o equivalentemente en su descripción discreta, de la forma $t \rightarrow n$, $x(t)\sim x(n)$, puede ser cambiada en el tiempo mediante la correspondencia:
### $n \rightarrow n-k$, con $k \in \mathbb{Z}$. 

### Si $k>0$ entonces hablamos de que la modificación temporal corresponde a una señal de tiempo retrasado por $k$ unidades de tiempo (time delay signal). Por otra parte, si $k<0$ hablamos de una modificación temporal correspondiente a una señal de tiempo adelantado por $|k|$ unidades de tiempo (time advance signal).

### Vamos a considerar que $k=\pm 2$ unidades temporales:


# $\blacksquare$ Señal impulso unitario de tiempo retrasado (*Time Delay Impulse*):

### Para $\delta(t)$ definimos que cuando $t=0$ la señal tiene su amplitud máxima (unitaria), por lo que cuando $k =+2$, vamos a definir una $\delta_{delay}=\delta_{delay}(t-2)$, teniendo una señal de tiempo retrasado por $2$ unidades temporales. Es decir, vamos a registrar la amplitud máxima dos unidades temporales después con respecto a $\delta(t)$ (cuando $t \rightarrow t+2$), definiendo así:

### $ \delta_{delay}(t_{delay}=t-2)= \begin{cases}
    1,& t_{delay}=t+2\\
    0,              & t_{delay} \neq 0
\end{cases}$

```python
impulso_delay=signal.unit_impulse(7, 2) #7 puntos temporales, el impulso en t=2
print(impulso)
plt.figure(figsize = (18,8)) 
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(impulso_delay,linewidth = 3,color="cyan")
plt.xlabel('time.', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.xlim([-2, 5])
plt.title(r"Impulso unitario de tiempo retrasado, $t \rightarrow t-2$:", fontsize = 40, color="yellow")
plt.xlabel('$t$', fontsize = 30,color ="yellow")
plt.ylabel('$\delta_{delay} (t)$, amplitud', fontsize = 30,color ="yellow")
plt.grid(True)
plt.show()
```

# $\blacksquare$ Señal de escalón unitario:

### Esta señal está caracterizada analíticamente por la función:

$ u(t)= \begin{cases}
    1,& \text{para } t\geq 0\\
    0,&   \text{si }            t < 0
\end{cases}$

### Tenemos el mismo caso de la delta de Dirac. Aunque hablando en términos del análisis real esta función no es continua (pues los límites por izquierda y por derecha no son iguales en cero), físicamente la señal sí lo es. Se toma que se presenta una pendiente creciente cuando $t \in [-1,0)$.

### Su visualización gráfica es la siguiente:

```python
t=[]
u=[]
for i in range(-6,10):
    t.append(i)
    if i >= 0:
        u.append(1)
    else:
        u.append(0)

print(t)
print(u)
```

```python
plt.figure(figsize = (18,8)) 
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,u,linewidth = 3,color="cyan")
plt.ylim(0,1.5)
plt.xlim(-5,6)
plt.title(r"Escalón unitario:", fontsize = 40, color="yellow")
plt.xlabel('$t$', fontsize = 30,color ="yellow")
plt.ylabel('$u(t)$, Amplitud', fontsize = 30,color ="yellow")
plt.grid(True,axis="y")
plt.show()
```

# $\blacksquare$ Señal de rampa unitaria:

### Representada mediante la función:

$ x(t)= \begin{cases}
    t,& \text{si } t\geq 0\\
    0,&   \text{si }            t < 0
\end{cases}$

```python
t=[]
x=[]
for i in range(-6,10):
    t.append(i)
    if i >= 0:
        x.append(i)
    else:
        x.append(0)

print(t)
print(x)
```

```python
plt.figure(figsize = (18,8))
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x, linewidth = 3, color="cyan")
plt.title(r"Rampa unitaria:", fontsize = 40, color="yellow")
plt.xlabel('$t$', fontsize = 30,color ="yellow")
plt.ylabel('$x(t)$, Amplitud', fontsize = 30,color ="yellow")
plt.grid(True)
plt.show()
```

# $\blacksquare$ Señal exponencial de amplitud unitaria:

### Descrita mediante la función: $x(t)=A \cdot e^{-t}$, para $0 \leq t < \infty$ y $A=1$.

### O bien, podemos definirla como:

$ x(t)= \begin{cases}
    1,& \text{si } t= 0\\
    e^{-t},&   \text{si }            t > 0
\end{cases}$

```python
t=np.linspace(0,10,100)
x=np.exp(-t)
```

```python
plt.figure(figsize = (15,8))
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x,linewidth = 3, color="cyan")
plt.title(r"Exponencial unitaria:", fontsize = 40, color="yellow")
plt.xlim(0,5)
plt.ylim(0,1)
plt.xlabel('$t$', fontsize = 30,color ="yellow")
plt.ylabel('$x(t)$, Amplitud', fontsize = 30,color ="yellow")
plt.grid(True)
plt.show()
```

# $\blacksquare$ Señal sinusoidal de fase cero:

### En este caso hablamos de una señal que presenta un comportamiento oscilatorio, descrita mediante la función:

## $y(t)=Asen(2 \pi \nu t)$

### En donde vamos a definir la amplitud $A=5$ unidades de distancia, digamos metros, y la frecuencia $\nu=50 Hz$.

```python
t=np.linspace(0,10,100)
A=5 #amplitud
f=50  #frecuencia
y=A*np.sin(2*np.pi*f*t)
```

```python
plt.figure(figsize = (15,8))
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,y,linewidth = 3,color="cyan" )
plt.xlim(0,10)
plt.ylim(-5,5)
plt.title(r"Onda sinusoidal:", fontsize = 40, color="yellow")
plt.xlabel('$t$', fontsize = 40, color="yellow")
plt.ylabel('$y(t), $Amplitud', fontsize = 40, color="yellow")
plt.grid(True)
plt.show()
```
