# Fundamentos matemáticos de la convolución:

### Recordemos que para dos funciones $f,h\in \mathbf{L_1}(\mathbb{R}):=\{g:\mathbb{R} \rightarrow \mathbb{R}|\int _{-\infty} ^{\infty}  |g(x)|dx < \infty \}$, la convolución de $f$ con $h$ se define como:

<font size="5"> $(f \diamond h)(t)=\int_{-\infty} ^{\infty} f(t-x)h(x)dx= \int_{-\infty} ^{\infty} f(x) h(t-x)dx$ </font>

### (Dependiendo de la literatura y del contexto, a veces esta expresión puede aparecer como $\frac{1}{\sqrt{2\pi}}\int_{-\infty} ^{\infty} f(t-x)h(x)dx$).

### Donde $f(t)$ es la función trasladada en $t$ unidades sobre $x$ con respecto al origen.

### El operador convolución cumple con ser lineal, conmutativo y asociativo.

### Cuando hablamos de señales analógicas, recordemos que estas pueden ser muestreadas, por lo que para describir el sistema de tiempo discreto resultante, una de las clasificaciones que podemos mencionar es la de tiempo invariante o tiempo variante.

### Hablando en el contexto de las matemáticas del procesamiento de señales, si tenemos un sistema $\mathbb{T}$ en un estado relajado (sin ninguna señal de entrada), podemos excitarlo con una señal de tiempo discreto de entrada $x(n)$; por lo que $\mathbb{T}$ produce una señal de salida $y(n)=\mathbb{T}[x(n)]$.
    
### Recordamos que se presenta un retraso (o puede ser un adelanto) temporal si $x(n) \rightarrow x(n-k), \forall k \in \mathbb{Z}$.

### Decimos que un sistema relajado $\mathbb{T}$ es invariante temporalmente si:

<font size="5"> $x(n) \xrightarrow{\mathbb{T}} y(n)$**, y entonces,** $x(n-k) \xrightarrow{\mathbb{T}} y(n-k)$, $\forall x(n),k$. </font>

### De lo anterior, en general podemos escribir la señal de salida como:

<font size="5"> $y(n,k)=\mathbb{T}[x(n-k)]=y(n-k)$ </font>

### Un sistema que no cumpla la definición anterior es clasificado como variante temporalmente.

# Sobre la resolución de señales de tiempo discreto en impulsos $x(n)$:

### Podemos suponer una señal de entrada sobre un sistema, la cual está resuelta en componentes de señal elementales ($\{x_k\}$), escrita de manera analítica como:

<font size= 5> $x(n)=\sum _k C_k x_k(n)$ </font>

### En donde $C_k$ es la amplitud de la componente $x_k (n)$.

### A la respuesta del sistema a cada componente de señal de entrada la definimos como:

<font size=5> $y_k(n):=\mathbb{T}[x_k(n)]$ </font>

### Con lo anterior podemos resolver $x(n)$ en una suma de secuencias de muestreo unitarias, considerando el producto $x_k(n)=\delta (n-k)$. En donde $\delta(n-k)$ es la muestra del pulso unitario a un retraso de $k$ unidades de tiempo. Entonces:

<font size=5> $x(n)\cdot \delta (n-k)=x(k)\cdot \delta (n-k)=
\begin{cases}
    x(k),& \text{si } n=k\\
    0,              & \text{si } n\neq k
\end{cases}$ </font>

### Si todos los retrasos posibles son $-\infty <k< \infty$, para la suma de todos los productos de secuencias es la resolución de una señal arbitraria de entrada en una suma de *shifted sample sequences*:

<font size=5> $x(n)=\sum _{k=-\infty} ^{\infty} x(k)\delta(n-k)$ </font>

### Ahora determinamos la respuesta $y(n,k)$ de un sistema $\mathbb{T}$ a $x(n)$.  Definamos:

<font size=5> $y(n,k):=h(n,k)=\mathbb{T}[\delta (n-k)]$, con $-\infty <k< \infty$ </font>

### El término $y(n,k)$ es la respuesta para un sistema lineal relajado ante la interacción con una señal de entrada arbitraria $x(n)$. Por lo que tenemos:

<font size=5> $y(n)=\mathbb{T}[x(n)]=\mathbb{T}[\sum _{k=-\infty} ^{\infty} x(k)\delta(n-k)]=\sum _{k=-\infty} ^{\infty} x(k)\mathbb{T}[\delta (n-k)]$ </font>

### (pues al ser $\mathbb{T}$ lineal, abre sumas y saca escalares), entonces: 

<font size=5> $y(n):=\sum _{k=-\infty} ^{\infty} x(k)h(n,k)$ </font>

### De lo mencionado anteriormente, podemos finalmente escribir la señal de entrada $x(n)$ convolucionada con la respuesta del sistema lineal discreto invariante en el tiempo (*Linear Time Invariant System, LTI*) al impulso unitario $\delta(n)$,  $h(n)$ como:

<font size=5> $y(n)=\sum _{k=-\infty} ^{\infty} x(k)h(n-k)$ </font>

### Definimos $(x \diamond h)(n)=y(n)$ como la suma de convolución. Notemos que con respecto a la convolución definida al principio del notebook, este es un equivalente discreto aplicado a señales.  

### Para calcular $y$ alrededor de un $n=n_0$, normalmente se definen cuatro pasos para calcular la convolución:

### 1) Folding (o flipping):  Se define el proceso $h(k) \rightarrow h(-k)$ (*foldear* $h(k)$ para obtener $h(-k)$).

### 2) Shifting: Con base al proceso $h(k) \rightarrow h(-k)$, se cambia (recorre, *shiftea*, traslada o cualquier traducción equivalente y apropiada) $h(-k)$ por $n_0$ unidades temporales. Si $n_0 \in \mathbb{Z}^+$ es un recorrido a la derecha.  Si $n_0 \in \mathbb{Z}^-$ se recorre a la izquierda.  

### 3) Multiplicación: Corresponde al proceso de los prouctos secuencia $x(k)\cdot h(n_0 -k)$

### 4) Suma: Corresponde al valor de la salida a tiempo discreto $n=n_0$.

# ¿Cómo tratamos la suma sobre valores infinitos $\sum _{k=-\infty} ^{\infty}$ en $y(n)$ numéricamente en Python?

### Recordando que en un sistema LTI, a un tiempo $n=n_0$, la salida está dada por la convolución  $y(n=n_0)=\sum _{k=-\infty} ^{\infty} x(k)h(n_0-k)=\sum _{k=-\infty} ^{\infty} h(k)x(n_0-k)$

### Podemos expresar la suma anterior en dos términos:

<font size=5> $\begin{cases}
    1) \text{ Valores pasados y presentes de la señal de entrada}:&  x(n), \text{ para } n\leq n_0 \\
    2)\text{ Valores futuros de la entrada}:& x(n), \text{ para } n>n_0
 \end{cases}$ </font>
 
 ### Es decir:
 
 <font size=5> $y(n_0)=\sum _{k=0} ^{\infty} h(k)x(n_0-k)+\sum _{k=-\infty} ^{-1} h(k)x(n_0-k)=[h(0)x(n_0)+h(1)x(n_0-1)+h(2)x(n_0-2)+...]+[h(-1)x(n_0+1)+h(-2)x(n_0+2)+...]$ </font>
 
 ### Decimos que un sistema es casual cuando la señal de salida $y(n)$ depende únicamente de valores presentes y pasados en la entrada, no de valores futuros.
 
 ### La causalidad del sistema puede traducirse bajo la siguiente condición para la respuesta al impulso:

<font size=5> $h(n)=0$, si $n<0$ </font>

### Por lo que para una respuesta de impulso de duración infinita a tiempo $n=n_0$ tenemos:

<font size=5> $y(n_0)=\sum _{k=0}^{\infty} h(k)x(n_0-k)$ </font>

### Si la respuesta de impulso es finita, entonces bajo la condición $h(n)=0$ para $n<0$ y $n\geq M$, tenemos la expresión:

<font size=5>  $y(n_0)=\sum _{k=0}^{M-1} h(k)x(n_0-k)=\sum _{k=0}^{L-1} x(k)h(n_0-k)$ </font>

### En donde $M$ es el número de muestras de la señal de entrada.

### A este modo de convolución se le conoce como *Full Mode Convolution*, en donde el número de muestras de la señal de salida está dado por:

<font size="5"> $N_{conv}=L+M-1$ </font>

### En donde $L$ es el número de muestras de $x(n)$ y $M$ el de $h(n)$. Entonces se halla el número de secuencias en la señal de salida usando la convolución a modo completo (hay más métodos, pero parece que este es el más importante y el más usado).

### Para la *full mode convolution* vamos a definir el siguiente comando:

```python
y=np.convolve(x,h,mode = 'full')
y
```

```python
plt.figure(figsize = (20,8)) 
plt.suptitle('Convolución a modo completo de $x(n)$ con $h(n)$', fontsize = 40, color="orange")
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 25
plt.plot(y, 'c-', linewidth =3,label = 'Convolución usando np.convolve a modo completo')
markerline, stemlines, baseline = plt.stem(y)
plt.setp(stemlines, 'linewidth', 3,color="r",label = 'Muestras de la señal de salida') 
plt.setp(baseline,color="white")
plt.setp(markerline,color="r")
plt.xlabel(r'Muestra, $n$', fontsize = 30, color="yellow")
plt.ylabel(r'Valor de la convolución, $y(n)$', fontsize = 30, color="yellow")
plt.legend(fontsize = 15)
plt.grid(True)
plt.show()
```
